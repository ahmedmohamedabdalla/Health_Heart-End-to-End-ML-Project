<h4>Building Model</h4>

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.metrics import plot_confusion_matrix
```

```python
features=['Age', 'Sysbp', 'Diabp', 'Hr', 'Weightkg', 'Heightcm', 'Bmi']
X=heart_data[features]
y=heart_data['Indication']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
lo=LogisticRegression(max_iter=1000)
lo.fit(X_train,y_train)
```

```python
dec=DecisionTreeClassifier(max_depth=30)
dec.fit(X_train,y_train)
```

```python
print('Score of Training of Logistic Regression {} %'.format(lo.score(X_train,y_train).round(5)))
print("\033[1m---------------------------------\033[0m")
print('Score of Testing of Logistic Regression {} %'.format(lo.score(X_train,y_train).round(5)))
print("\033[1m---------------------------------\033[0m")
c=cross_val_score(lo,X_train,y_train).round(3)
cr=pd.Series(c)
print('Cross Validation Score \n{}'.format(cr))
print("\033[1m---------------------------------\033[0m")
print(accuracy_score(y_test,lo.predict(X_test)))
```

```python
print('Score of Training of Decision Tree {} %'.format(dec.score(X_train,y_train).round(7)))
print("\033[1m---------------------------------\033[0m")
print('Score of Testing of Decision Tree {} %'.format(dec.score(X_train,y_train).round(7)))
print("\033[1m---------------------------------\033[0m")
c=cross_val_score(dec,X_train,y_train).round(3)
cr=pd.Series(c)
print('Cross Validation Score \n{}'.format(cr))
print("\033[1m---------------------------------\033[0m")
print(accuracy_score(y_test,dec.predict(X_test)))
```

```python
co=confusion_matrix(y_test,lo.predict(X_test))
co_1=confusion_matrix(y_test,dec.predict(X_test))
```

```python
plt.figure(figsize=(12,5))
plt.suptitle('Confusion_Matrix of Logistic Regression & Decision Tree',fontsize=20)
plt.subplot(1,2,1)
sns.heatmap(co,annot=True)
plt.subplot(1,2,2)
sns.heatmap(co_1,annot=True)
plt.show()
```

```python
print('Classification Report for Logistic Regression')
print("\033[1m-----------------------------------------------------\033[1m")
print(classification_report(y_test,lo.predict(X_test)))
```

```python
print('Classification Report for Decision Tree')
print("\033[1m-----------------------------------------------------\033[1m")
print(classification_report(y_test,lo.predict(X_test)))
```
